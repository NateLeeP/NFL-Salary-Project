```python
""" For purposes of analysis only interested in quarterbacks who started more than half (9 or greater) of their teams games. Position must be QB"""
```

```python
nfl_qb_stats = nfl_raw_stats_df[nfl_raw_stats_df['Pos'] == 'QB']
```

```python
#Dataframe is limited to QBs who started at least half of their teams games
#Loop prints out series of each column, double checking numerical stats are read in as floats
for column in nfl_qb_stats.columns:
    print(nfl_qb_stats[column])
```

```python
#Break QB record into 'Wins' and 'Losses'

def wins_and_losses(row):
    # Have to return row!!! Or get none
    wins = row['QBrec'].split('-')[0]
    losses = row['QBrec'].split('-')[1]
    ties = row['QBrec'].split('-')[2]
    row['Wins'] = wins
    row['Losses'] = losses
    row['Ties'] = ties
    return row
    
```

```python
#Create columns for Wins, Losses, and Ties
nfl_qb_stats.loc[:,'Wins'] = [0 for x in range(len(nfl_qb_stats))]
nfl_qb_stats.loc[:,'Losses'] = [0 for x in range(len(nfl_qb_stats))]
nfl_qb_stats.loc[:,'Ties'] = [0 for x in range(len(nfl_qb_stats))]
```

```python
# Use Wins and Losses, as easier to use as feature. 
nfl_qb_stats = nfl_qb_stats.apply(lambda x: wins_and_losses(x), axis = 1)
```

```python
# Clean up player name, remove * and +
nfl_qb_stats['Player'] = nfl_qb_stats['Player'].apply(lambda x: x.strip('+').strip('*'))
```

```python
## Next Step: Clean dataframe of player salaries
nfl_qb_salaries = pd.read_csv(r'C:\Users\nlpru\Desktop\NFL Salary Project\NFL Project CSV\nfl_salaries_raw_data_pd.csv')
```

```python
# Drop unnecessary columns. Change column names. 
# Running this again will cause an ERROR
nfl_qb_salaries = nfl_qb_salaries.drop(columns = nfl_qb_salaries.columns[:2].values).rename(columns = {' average':'Salary'})
```

```python
# Function to clean player name. 
def clean_player_name(player):
    name = player.split('\n')[1]
    return name
nfl_qb_salaries['Player'] = nfl_qb_salaries['Player'].apply(clean_player_name)
```

```python
#Function to convert salary into an int
def convert_sal_to_int(salary):
    salary = int(salary.strip('$').replace(',', ''))
    return salary
nfl_qb_salaries['Salary'] = nfl_qb_salaries['Salary'].apply(convert_sal_to_int)
```

```python
#Merge datasets together
nfl_stats_salary = pd.merge(nfl_qb_stats, nfl_qb_salaries, how = 'left', on = 'Player')
# Merge works, but need to fix QB names in NFL stats
```

```python
def change_na(stat):
    if pd.isna(stat):
        return 0
    else:
        return stat
```

```python
# Removes the NaN from Game-winning Drive and 4th Quarter Comeback. NaN represented 0. 
nfl_stats_salary['GWD'] = nfl_stats_salary['GWD'].apply(change_na)
nfl_stats_salary['4QC'] = nfl_stats_salary['4QC'].apply(change_na)

# Dropping QBR (blank column)
nfl_stats_salary.drop(columns = ['QBR'], inplace = True)
```

```python
# Save 'nfl_stats_salary' as a csv for Linear Regression analysis
nfl_stats_salary.to_csv(r'C:\Users\nlpru\Desktop\NFL Salary Project\NFL Project CSV\nfl_stats_salary.csv', index = False)
```
